```python
#| export
from typing import List, Optional, Protocol, runtime_checkable

from cjm_file_discovery.core.models import FileInfo
from cjm_file_discovery.core.config import ScanConfig
```

## DiscoveryProvider Protocol

Abstract interface for file discovery backends. Implement this protocol to support different file systems (local, S3, Google Drive, etc.).

```python
#| export
@runtime_checkable
class DiscoveryProvider(Protocol):
    """Protocol for file discovery backends."""

    @property
    def name(self) -> str:  # Unique identifier for this provider
        """Unique identifier for this provider (e.g., 'local', 's3')."""
        ...

    def scan(
        self,
        directories: List[str],  # Directories to scan
        config: ScanConfig        # Scan configuration
    ) -> List[FileInfo]:  # List of discovered files
        """Scan directories for files matching config."""
        ...

    async def scan_async(
        self,
        directories: List[str],  # Directories to scan
        config: ScanConfig        # Scan configuration
    ) -> List[FileInfo]:  # List of discovered files
        """Async scan for files."""
        ...

    def get_file_info(
        self,
        path: str  # Path to file
    ) -> Optional[FileInfo]:  # FileInfo or None if not found
        """Get metadata for a single file."""
        ...

    def supports_path(
        self,
        path: str  # Path to check
    ) -> bool:  # True if this provider can handle the path
        """Check if this provider can handle the given path."""
        ...
```

```python
# Test that protocol is runtime checkable
class MockProvider:
    """Mock provider for testing protocol."""
    @property
    def name(self) -> str:
        return "mock"
    
    def scan(self, directories, config):
        return []
    
    async def scan_async(self, directories, config):
        return []
    
    def get_file_info(self, path):
        return None
    
    def supports_path(self, path):
        return True

# Protocol check works
mock = MockProvider()
assert isinstance(mock, DiscoveryProvider)

# Non-implementing class fails check
class NotAProvider:
    pass

assert not isinstance(NotAProvider(), DiscoveryProvider)

print("DiscoveryProvider protocol tests passed!")
```
